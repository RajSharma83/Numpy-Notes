```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.random.rand(5)
arr3 = np.ones((3, 3))
```

```python
np.save('array1.npy', arr1)
np.save('array2.npy', arr2)
np.save('array3.npy', arr3)
```

```python
load_arr1 = np.load('array1.npy')
print(load_arr1)
load_arr2 = np.load('array2.npy') 
print(load_arr2)  
load_arr3 = np.load('array3.npy')
print(load_arr3)
```

```python
try:
    logo = np.load('numpy.svg.npy')
    #Display
    plt.Figure(figsize=(5,5))
    plt.Subplot(121)
    plt.show(logo)
    plt.title('NumPy Logo Image')
    plt.grid(False)

    dark_logo = 1 - logo
    plt.Subplot(122)
    plt.show(dark_logo)
    plt.title('Dark NumPy Logo Image')
    plt.grid(False)
except FileNotFoundError:
    print("File not found. Please check the file path and try again.")
```
