```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
 # Data Structure of Resturants : [resturannt_id , 2022 , 2023 , 2024 , 2025]

data = np.array([
    [1, 5000, 5500, 6000, 6500], # Star Biryani
    [2, 7000, 7200, 7400, 7600], # Ekta Resturant
    [3, 80000, 8500, 9000, 9500],  # Barbeque Nation
    [4, 6000, 6200, 6400, 6600],  # Spice Hub
    [5, 9000, 9500, 10000, 10500] # Curry Palace
])

print("----- Zomato Data Analysis -----")
print("Data Shape (Resturants x Years):", data.shape)
print("Data Type:", data.dtype)
print("Data Dimensions:", data.ndim)
print("Total Elements in Data:", data.size)
print("Resturant 0 to 2 Data:\n", data[0:3])
```

```python
print("\n--- Revenue Analysis ---")
print(np.sum(data, axis= 0)) # Total revenue for each year including resturant id  column
print(np.sum(data, axis= 1))   # Total revenue for each resturant including year id row
year_tot = (np.sum(data[:,1:], axis= 0))  
print("Total Revenue for each Year:", year_tot)

# Total revenue for each resturant over the years
total_revenue = np.sum(data[:, 1:], axis=1)
print("Total Revenue for each Resturant:", total_revenue)
```

```python
# Minium sale per resturant 
min_revenue = np.min(data[:,1:], axis=1)
print("Minimum Revenue for each Resturant:", min_revenue)

# Maximum sale per  years
max_revenue = np.max(data[:,1:], axis=1)
print("Maximum Revenue for each Resturant:", max_revenue)

# Average sale per resturant
avg_revenue = np.mean(data[:,1:], axis=1)
print("Average Revenue for each resturant:",avg_revenue)
```

```python
cusum = np.cumsum(data[:,1:], axis=1)
print("Cumulative sum of revenue for each resturant over the years:\n", cusum)

plt.figure(figsize=(8, 8))
plt.plot(np.mean(cusum, axis=0))
plt.title("Average Cumulative Revenue Over Years")
plt.xlabel("Years (2022-2025)")
plt.ylabel("Average Cumulative Revenue")
plt.grid()
plt.show()
```

```python
vecotr1 = np.array([1, 2, 3])
vecotr2 = np.array([4, 5, 6])

print("Vector1 and Vector2 addition:", vecotr1 + vecotr2)
print("Vector1 and Vector2 multiplication:", vecotr1 * vecotr2)
print("Vector1 and Vector2 dot product:", np.dot(vecotr1, vecotr2))

angle = np.arccos(np.dot(vecotr1, vecotr2) / (np.linalg.norm(vecotr1) * np.linalg.norm(vecotr2)))
print("Angle between Vector1 and Vector2 (in radians):", angle)
```
